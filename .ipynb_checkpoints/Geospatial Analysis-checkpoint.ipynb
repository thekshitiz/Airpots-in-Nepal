```python
# importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
# reading data from csv

df = pd.read_csv("Data/airport-list.csv")
df.tail(10)
```

## create a scatter plot

```python
# Create a scatter plot

plt.figure(figsize=(10, 6))  # Set the size of the figure
plt.scatter(df['longitude_deg'], df['latitude_deg'], s=30, alpha=0.5)  # Adjust 's' to change the size of the points
# Customize the plot
plt.title('Geospatial Distribution')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Display the plot
plt.show()
```

```python
# Calculate the count of items by continent or country
count_by_region = df.groupby('continent')['id'].count()  # Replace 'continent' with 'iso_country' if needed

# Load the shapefile for continents or countries
shapefile_path = 'Data/shapefile.shp'  # Replace with the actual path to the shapefile

# Read the shapefile using geopandas
gdf = gpd.read_file(shapefile_path)

# Merge the count data with the shapefile data
merged_data = gdf.merge(count_by_region, left_on='continent_column_name', right_index=True, how='left')

# Create a choropleth map
plt.figure(figsize=(10, 6))  # Set the size of the figure
merged_data.plot(column='id', cmap='YlOrRd', linewidth=0.8, edgecolor='0.8', legend=True)  # Adjust the colormap (cmap) as needed

# Customize the plot
plt.title('Choropleth Map - Number of Items by Continent')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Display the plot
plt.show()
```
